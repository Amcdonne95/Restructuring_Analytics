```python
df = pd.read_csv('DIPS_Data.csv')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df.head()
```

```python
df = df[~df.case_name.isin(['Fees:', 'Agents:','Lender Groups:', 'Roll ups:', 'Lenders:'])]
```

```python
df.head()
```

```python
df = df.dropna(subset=['case_name'])
```

```python
df.head()
```

```python
df_sub = df[['case_name','facility_type', 'fixed/floating', 'investigation_budget','carve_out', 'deadline', 'priotity/prime_everything',
            'collateral_all_assets', 'professional_line_item']]
```

```python
df_sub.head()
```

```python
df_missing = df_sub[pd.isnull(df_sub).any(axis=1)]
```

```python
df_missing
```

```python
file_name = 'Missing_Figures.csv'
df_missing.to_csv(file_name, encoding='utf-8', index=False)
```
