```python
df = pd.read_csv('DIPS_Data.csv')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
```

```python
df_sub = df.copy()
```

```python
value = 'Fees:'
case = 'xxxx'
for row in df_sub.itertuples():
    if row[1] not in ['none', 'Fees:', 'Agents:','Lender Groups:', 'Roll ups:', 'Lenders:']:
        case = row[1]
        df_sub.at[row[0], 'Case'] = case
        continue
    if row[1] == 'Fees:':
        value = 'Fees:'
        continue
    if row[1] == 'Agents:':
        value = 'Agents:'
        continue
    if row[1] == 'Lender Groups:':
        value = 'Lender Groups:'
        continue
    if row[1] == 'Roll ups:':
        value = 'Roll ups:'
        continue
    if row[1] == 'Lenders:':
        value = 'Lenders:'
        continue
    df_sub.at[row[0], 'Case'] = case
    df_sub.at[row[0], 'case_name'] = value
```
